1. <b>Filter Methods: </b>In filter method a subset of features are selected from the available set of features.
2. <b>Wrapper Methods: </b>
    - <b>Forward Selection: </b>Forward selection is an iterative method in which we start with having no feature in the model. In each iteration, we keep adding the feature which best improves our model till an addition of a new variable does not improve the performance of the model.
    - <b>Backward Elimination: </b>In backward elimination, we start with all the features and removes the least significant feature at each iteration which improves the performance of the model. We repeat this until no improvement is observed on removal of features.
3. <b>Embedded Methods: </b>In embeedded method all possible subset of datasets are created from the dataset and models are built on each subset of data and metrics are calculated. The subset of features which gives the best metric are considered as best features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Univariate Selection
- Statistical tests can be used to select those features that are strongly related to the dependent feature.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
df = pd.read_csv("mobile.csv")
df.head()
```

```python
X = df.iloc[:,0:20]  # independent columns
y = df.iloc[:,-1]    # target column i.e price range
```

```python
# apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
```

```python
feature_scores = pd.concat([pd.DataFrame(X.columns), pd.DataFrame(fit.scores_)], axis = 1)
feature_scores.columns = ["feature", "score"]
feature_scores.sort_values(by="score", ascending=False)
```

```python
feature_scores.nlargest(10, "score")
```

## Feature Importance
- We can get the feature importance of our dataset using the feature_importance_ property of our model.

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
plt.figure(figsize=(14,8))
plt.barh(X.columns, model.feature_importances_)
plt.title("Feature Importances")
plt.show()
```
